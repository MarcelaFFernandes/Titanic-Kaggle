```python
train = pd.read_csv("//kaggle/input/titanic/train.csv")
test = pd.read_csv("/kaggle/input/titanic/test.csv")


test.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
modelo = RandomForestClassifier(n_estimators=100, random_state=0)
```

```python
variaveis = ["Sexo_binario", "Age"]
train["Sex"].value_counts()
```

```python
def sex_binario(valor):
    if valor == 'female':
        return 1 
    else: 
        return 0
```

```python
train["Sexo_binario"] = train["Sex"].map(sex_binario)
train.head()
```

```python
x = train[variaveis]
y = train["Survived"]
x.head()
```

```python
y.head()
```

```python
train.head()
```

```python
test.head()
```

```python
x = x.fillna(-1)
modelo.fit(x, y)
```

```python
test.head()
```

```python
test["Sexo_binario"] = test['Sex'].map(sex_binario)
test["Age"] = train["Age"]
test.head()
```

```python
x_prev = test[variaveis]
x_prev = x_prev.fillna(-1)
x_prev.head()
```

```python
p = modelo.predict(x_prev)
p
```

```python
test.head()
```

```python
#transformando array em série de pandas com index PassengerId
resultado = pd.Series(p, index = test["PassengerId"], name = "Survived")
```

```python
#criando arquivo csv
resultado.to_csv("modelo_1.csv", header = True)
```
